```python
class FalseLLMChain:

    def __init__(self, llm, prompt):
        self.llm = llm
        self.prompt = prompt

    def run(self, input_dic):
        final_prompt = self.prompt.format(input_dic)
        result = self.llm.predict(final_prompt)

        return result['response']
```
